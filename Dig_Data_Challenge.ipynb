Episode rate of completion: Let's compute the rate at which each episodes are watched. Do the viewers complete it? This will help us know how successful an episode is. This will be calculated as

Rate of completion = (Total Episode watch seconds/Total movie session seconds) * 100
But first, lets group by the user id and the episode and sum the watch time. This is because the viewers can complete the movie in two sessions

```python
spy_amongst_friends = itv_data.loc[(itv_data['programme_title'] == 'A SPY AMONG FRIENDS')].reset_index()
spy_amongst_friends = spy_amongst_friends.drop(columns=['index'])
spy_amongst_friends
```

```python
spy_amongst_friends = spy_amongst_friends[['user_id', 'episode_number','consumption_seconds']]
spy_amongst_friends
```

Now to calculate the watch rate, First, I summed the total consumption seconds of each viewer per episode.

```python
spy_amongst_friends_sum = spy_amongst_friends.groupby(['user_id','episode_number'])['consumption_seconds'].sum().reset_index()
spy_amongst_friends_sum
```

I browsed the net to find out how long each episodes are then converted them to seconds. Here, I created a new column containing the episode total seconds

```python
def categorise4(row):
  if row['episode_number'] == 1:
    return 3300
  elif row['episode_number'] == 2:
    return 3600
  elif row['episode_number'] == 3:
    return 3240
  elif row['episode_number'] == 4:
    return 3300
  elif row['episode_number'] == 5:
    return 3120
  elif row['episode_number'] == 6:
    return 3480

spy_amongst_friends_sum["espisode_total_seconds"] = spy_amongst_friends_sum.apply(lambda row: categorise4(row), axis=1)
spy_amongst_friends_sum
```

Calculating the watch rate of the viewers. This is calculated by dividing the viewers total consumption time by the episode total seconds and multiplying by a 100%.

```python
spy_amongst_friends_sum['watch_rate'] = (spy_amongst_friends_sum.consumption_seconds/spy_amongst_friends_sum.espisode_total_seconds) * 100
spy_amongst_friends_sum
```

```python
watch_rate = spy_amongst_friends_sum.groupby(['episode_number'])['watch_rate'].mean().reset_index()
watch_rate
```

You'd notice that the total watch consumption time of some of the viewers is double times the episode consumption time. This is because the viewer watched the episode again another day. These viewers are the repeat watchers. 

```python
def categorise4(row):
  if row['watch_rate'] > 100:
    return 1
  else:
    return 0

spy_amongst_friends_sum["repeat_watchers"] = spy_amongst_friends_sum.apply(lambda row: categorise4(row), axis=1)
spy_amongst_friends_sum
```
